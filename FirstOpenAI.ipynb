ChatOpenAI를 임포트 한다.

```python
from langchain_openai import ChatOpenAI
```

OPENAI 키는 비밀리에 다뤄야 하므로, 외부에서 입력받도록 한다.<BR> 이를 위해 외부입력을 받는 모듈을 임포트 한다.

```python
import getpass
```

키를 외부로부터 받아, OPENAI_API_KEY 변수에 둔다.

```python
OPENAI_API_KEY = getpass.getpass("YOUR KEY: ")
```

실험용이므로, 성능에 상관없이 가장 저렴한 모델을 호출


```python
llm = ChatOpenAI(openai_api_key=OPENAI_API_KEY, model_name="gpt-5-nano")
```

프롬프트를 작성해 본다.

```python
prompt_input = """Tell me 10 topmost Univertsity in the world"""
```

엔진을 호출하고, 결과를 받는다.

```python
response = llm.invoke(prompt_input)
```

결과를 출력해 본다.

```python
print(response.content)
```

이번에는 명령 템플릿을 사용하여 호출해 본다.<BR> 먼저 프롬프트 템플릿을 임포트 한다.

```python
from langchain_core.prompts import PromptTemplate
```

템플릿을 작성한다.

```python
prompt_template = PromptTemplate.from_template("""You are an
experienced reporter. Write a {num_words} answer to the following question,
using a {tone} tone: {text}""")
```

질문 프롬프트를 작성한다.

```python
prompt_text = "Tell me 10 topmost beautiful cities in South Korea"
```

템플릿을 구성한다.

```python
prompt_input = prompt_template.format(
    text=prompt_text,
    num_words=100,
    tone="knowledgeable and engaging")
```

엔진을 호출하고, 결과를 받는다.

```python
response = llm.invoke(prompt_input)
```

결과를 출력해 본다.

```python
print(response.content )
```
