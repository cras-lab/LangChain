```python
import chromadb
chroma_client = chromadb.Client()
```

```python
docu_collection = chroma_client.create_collection(name="docu_collection")
```

```python
docu_collection.add(
    documents=[
        "주식 수익률의 로그값은 정규분포를 한다",
        "내일의 날씨는 매우 맑다.",
        "주식의 수익률은 두 수치의 비율로 구한다."
    ],
    metadatas=[
      {"source": "STOCK1"},
      {"source": "WEATHER"},
      {"source": "STOCK2"}
    ],
    ids=["ST-01", "W-01", "ST-02"]
)
```

Chroma는 코사인 유사도 사용해서 다음을 계산<BR>
유사도 = 1 - <코사인 유사도><BR>
0이 완전동일, 1은 무관, 2는 반대

```python
results = docu_collection.query(
    query_texts=["주식 정보 사이트 추천해 줘"],
    n_results=1
)

print(results)
```

```python
results = docu_collection.query(
    query_texts=["날씨는 정말 예측불가야"],
    n_results=1
)

print(results)
```

```python
results = docu_collection.query(
    query_texts=["날씨도 그런데, 주식종목이나 알려줘"],
    n_results=1
)

print(results)
```
