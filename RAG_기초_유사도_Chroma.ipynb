chromadb 클라이언트 생성: 실험을 위해 in-memory로 테스트

```python
import chromadb
chroma_client = chromadb.Client()
```

Chroma collection은 벡터 데이터 묶음(테이블/인덱스)이다.

```python
docu_collection = chroma_client.create_collection(name="docu_collection")
```

콜렉션에 값을 넣는다. BOW에 사용했던 값을 넣어서 콜렉션을 구축한다.

```python
docu_collection.add(
    documents=[
        "주식 수익률의 로그값은 정규분포를 한다",
        "내일의 날씨는 매우 맑다.",
        "주식의 수익률은 두 수치의 비율로 구한다."
    ],
    metadatas=[
      {"source": "STOCK1"},
      {"source": "WEATHER"},
      {"source": "STOCK2"}
    ],
    ids=["ST-01", "W-01", "ST-02"]
)
```

Chroma는 코사인 유사도 사용해서 다음을 계산<BR>
유사도 = 1 - <코사인 유사도><BR>
0이 완전동일, 1은 무관, 2는 반대

먼저 Q1을 해본다.

```python
results = docu_collection.query(
    query_texts=["주식 정보 사이트 추천해 줘"],
    n_results=1
)

print(results)
```

이제 날씨 관련 문구를 넣어 본다.

```python
results = docu_collection.query(
    query_texts=["내일 제주도 날씨 좀 찾아봐"],
    n_results=1
)

print(results)
```

날씨와 주식관련 단어를 모두 넣어서 결과가 어떤지 본다.

```python
results = docu_collection.query(
    query_texts=["날씨도 안 좋은데, 주식종목이나 알려줘.  "],
    n_results=1
)

print(results)
```
