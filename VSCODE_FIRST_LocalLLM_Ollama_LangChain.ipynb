langchain의 ollama 모듈을 임포트 한다.

```python
from langchain_ollama import ChatOllama
```

LLM을 생성한다.

```python
llm = ChatOllama(
    model="gemma3:4b",
    temperature=0
)
```

기본 프롬프트 골격을 만든다.

```python
messages = [
    ("system", "너는 한국어로 짧고 쉽게 답하는 AI 어시스턴트다.")
]
```

이제 루프를 돌며 LLM과 통신한다.

```python
print("Ollama + LangChain 로컬 챗봇 시작")
print("종료하려면 exit 입력")

while True:
    user_input = input("\n입력: ").strip()

    if user_input.lower() == "exit":
        print("종료합니다.")
        break

    if user_input == "":
        print("빈 입력이므로 종료합니다.")
        break

    print("\n사용자:", user_input)

    messages.append(
        ("human", user_input)
    )

    response = llm.invoke(messages)

    print("AI:", response.content)

    messages.append(
        ("ai", response.content)
    )
```
